# 確率勾配降下法

```python
# サンプルとする関数
#yの値を予想するAI

def f(x):
    y = 3 * x[0] + 2 * x[1]
    return y

# 初期設定
def init_network():
    # print("##### ネットワークの初期化 #####")
    network = {}
    nodesNum = 10
    network['W1'] = np.random.randn(2, nodesNum)
    network['W2'] = np.random.randn(nodesNum)
    network['b1'] = np.random.randn(nodesNum)
    network['b2'] = np.random.randn()

    # print_vec("重み1", network['W1'])
    # print_vec("重み2", network['W2'])
    # print_vec("バイアス1", network['b1'])
    # print_vec("バイアス2", network['b2'])

    return network

# 順伝播
def forward(network, x, mode):
    # print("##### 順伝播開始 #####")
    
    W1, W2 = network['W1'], network['W2']
    b1, b2 = network['b1'], network['b2']
    u1 = np.dot(x, W1) + b1
    z1 = None
    if mode==0:
      z1 = functions.relu(u1)
    else:
      ## 試してみよう
      z1 = functions.sigmoid(u1)
    
    u2 = np.dot(z1, W2) + b2
    y = u2

    # print_vec("総入力1", u1)
    # print_vec("中間層出力1", z1)
    # print_vec("総入力2", u2)
    # print_vec("出力1", y)
    # print("出力合計: " + str(np.sum(y)))    
    
    return z1, y

# 誤差逆伝播
def backward(x, d, z1, y, network):
    # print("\n##### 誤差逆伝播開始 #####")    

    grad = {}
    
    W1, W2 = network['W1'], network['W2']
    b1, b2 = network['b1'], network['b2']

    # 出力層でのデルタ
    delta2 = functions.d_mean_squared_error(d, y)
    # b2の勾配
    grad['b2'] = np.sum(delta2, axis=0)
    # W2の勾配
    grad['W2'] = np.dot(z1.T, delta2)
    # 中間層でのデルタ
    #delta1 = np.dot(delta2, W2.T) * functions.d_relu(z1)

    ## 試してみよう
    delta1 = np.dot(delta2, W2.T) * functions.d_sigmoid(z1)

    delta1 = delta1[np.newaxis, :]
    # b1の勾配
    grad['b1'] = np.sum(delta1, axis=0)
    x = x[np.newaxis, :]
    # W1の勾配
    grad['W1'] = np.dot(x.T, delta1)
    
    # print_vec("偏微分_重み1", grad["W1"])
    # print_vec("偏微分_重み2", grad["W2"])
    # print_vec("偏微分_バイアス1", grad["b1"])
    # print_vec("偏微分_バイアス2", grad["b2"])

    return grad


def stochastic_gradient_descent(learning_rate=0.01,mode=(0,0)):
  print("学習率{}".format(learning_rate))

  # サンプルデータを作成
  np.random.seed(seed=0)

  data_sets_size = 100000
  data_sets = [0 for i in range(data_sets_size)]
  if mode[0] == 0:print("初期値小")
  else: print("初期値大")

  for i in range(data_sets_size):
      data_sets[i] = {}
      if mode[0] == 0:
        # ランダムな値を設定
        data_sets[i]['x'] = np.random.rand(2)
      else:
        # 試してみよう_入力値の設定
        data_sets[i]['x'] = np.random.rand(2) * 2 - 1 # -5〜5のランダム数値(だとOverflowっぽい振る舞いなのでレンジを狭くする)
      
      # 目標出力を設定
      data_sets[i]['d'] = f(data_sets[i]['x'])
  
  losses = []
  # 学習率は引数で与える
  #learning_rate = 0.07

  # 抽出数
  epoch = 100000

  # パラメータの初期化
  network = init_network()
  # データのランダム抽出
  random_datasets = np.random.choice(data_sets, epoch)

  if mode[1] == 0:
    print("活性化関数はReLU")
  else:
    print("活性化関数はsigmoid")

  # 勾配降下の繰り返し
  for dataset in random_datasets:
      x, d = dataset['x'], dataset['d']
      z1, y = forward(network, x, mode[1])
      grad = backward(x, d, z1, y, network)
      # パラメータに勾配適用
      for key in ('W1', 'W2', 'b1', 'b2'):
          network[key]  -= learning_rate * grad[key]

      # 誤差
      loss = functions.mean_squared_error(d, y)
      losses.append(loss)
      #print(loss)

  #print("##### 結果表示 #####")    
  lists = range(epoch)

  plt.plot(lists, losses, '.')
  plt.ylim(0,1)
  # グラフの表示
  plt.show()
```

```python
#課題として「やってみよう」をやってみてそれをプロットして比較する
#そのため勾配降下法の部分も関数化して引数でやってみようを切り替える
#ついでに学習率も弄って収束具合を比べる
#エポック数は適当にダラダラとのばしてみた
lr=0.005
stochastic_gradient_descent(learning_rate=lr,mode=(0,0))
stochastic_gradient_descent(learning_rate=lr,mode=(1,0))
stochastic_gradient_descent(learning_rate=lr,mode=(0,1))
stochastic_gradient_descent(learning_rate=lr,mode=(1,1))

#まぁlossは徐々に減るがReLUとsigmoidの相違が強く表れる組み合わせを残す(それに意味があるかどうかは不明)
#初期値大かつReLUにて発散気味の挙動(学習率を下げると安定(下のCell参照))
#初期値はバラケすぎるとOverflowするといわれるので適当に抑えている(学習率も同じく)
#もしかしたら過学習気味?(まぁテストデータを作っての評価はしてないのでわからないけど)
```

```python
lr=0.001
stochastic_gradient_descent(learning_rate=lr,mode=(0,0))
stochastic_gradient_descent(learning_rate=lr,mode=(1,0))
stochastic_gradient_descent(learning_rate=lr,mode=(0,1))
stochastic_gradient_descent(learning_rate=lr,mode=(1,1))
#ReLUとsigmoidの差が逆転する？よくわからない
#学習率を大きい方にいじるとOverflowするのでやめておく
#学習率低くして安定方向だけど収束が甘いように見える
```
